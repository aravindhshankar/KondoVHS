```python
import numpy as np 
from matplotlib import pyplot as plt
```

### parameter $\rho_0 J_0 = x $
 
$$ T_K^M = D e^{-{\frac{1}{\rho_0 J_0}}}$$
$$ T_K^L = D e^{-\left(\frac{2}{\rho_0 J_0}\right)^{\frac{1}{2}}}$$
$$ T_K^H = \frac{D}{\left(1+\frac{\alpha D^\alpha}{\rho_0 J_0}\right)^{\frac{1}{\alpha}}} $$

```python
x1 = np.linspace(1e-2,0.3,1000)
D = 1 #We assume half bandwidth normalized this way 
alpha = 0.25

fig, (ax1,ax2) = plt.subplots(2)
TM1 = 1 * np.exp(-1/x1)
TL1 = 1 * np.exp(-(np.sqrt(2/x1)))
TH1 = 1 / (1 + (alpha* (D**alpha))/((1-alpha)*x1))**(1/alpha)

ax1.plot(x1,TM1,label = 'metal')
ax1.plot(x1,TL1, label = 'log vHS', ls = '-.')
ax1.plot(x1,TH1, label = 'HOvHS', ls = '--')
ax1.legend()
#plt.xscale('log')
ax1.set_xlabel(r'$\rho_0 \, J_0$')
ax1.set_ylabel(r'$T_K\,/D$')

x2 = np.logspace(-4,-2,1000)
TM2 = 1 * np.exp(-1/x2)
TL2 = 1 * np.exp(-(np.sqrt(2/x2)))
TH2 = 1 / (1 + (alpha* (D**alpha))/((1-alpha)*x2))**(1/alpha)

#ax2.plot(x2,TM2,label = 'metal')
ax2.plot(x2,TL2, label = 'log vHS', ls = '-.',c = 'tab:orange')
ax2.plot(x2,TH2, label = 'HOvHS', ls = '--', c ='tab:green')
ax2.legend()
ax2.set_yscale('log')
ax2.set_xlabel(r'$\rho_0 \, J_0$')
ax2.set_ylabel(r'$T_K \,/D$')
ax2.set_ylim(1e-12,1e-6)

plt.savefig('FigsPaper/ScalingComp.pdf')
```
